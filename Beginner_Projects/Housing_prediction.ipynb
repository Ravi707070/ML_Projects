```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r'C:\Dataset\PROJECTS\Housing.csv')

dataset = dataset.drop(columns=['mainroad','guestroom', 'basement', 'hotwaterheating','airconditioning','parking','prefarea', 'furnishingstatus'], axis=1)
dataset.head(3)
```

```python
sns.scatterplot(x='area', y='price', data=dataset, color='red')
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x = dataset.iloc[:,:-1]
y = dataset['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict([[7420, 4, 2, 3]])
```

```python
from sklearn.linear_model import  LogisticRegression
```

```python
lor = LogisticRegression(max_iter=1000)
lor.fit(x_train, y_train)
```

```python
lor.score(x_test, y_test)*10000
```

```python
print("Price",lor.predict([[7420, 4, 2, 3]]))
```
