```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('C:\Dataset\PROJECTS\Titanic-Dataset.csv')
dataset.drop(columns=['PassengerId','Fare','Cabin', 'Name', 'Sex', 'SibSp', 'Parch', 'Ticket'], inplace=True)
dataset.head(3)
```

```python
sns.scatterplot(x='Embarked', y='Pclass', hue='Survived', data=dataset)
plt.show()
```

```python
dataset.isnull().sum()
```

```python
dataset["Age"].fillna(dataset["Age"].mean(), inplace=True)
dataset["Embarked"].fillna(dataset["Embarked"].mode()[0], inplace=True)
```

```python
x = dataset.iloc[:,:-1]
y = dataset['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
```

```python
gb = GaussianNB()
gb.fit(x_train, y_train)
```

```python
gb.score(x_test, y_test)*100
```

```python
print("Surived", gb.predict([[1,3,22]]))
```

```python
mnb = MultinomialNB()
mnb.fit(x_train, y_train)
```

```python
mnb.score(x_test, y_test)*100
```

```python
print("Surived",mnb.predict([[0,3,22]]))
```

```python
bnb = BernoulliNB()
bnb.fit(x_train, y_train)
```

```python
bnb.score(x_test, y_test)*100
```

```python
print("Surived",bnb.predict([[0,2,12]]))
```
